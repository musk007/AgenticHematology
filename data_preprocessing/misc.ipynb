```python
## checking contents of images across diefferent folders for same magnification and same setting
import os
from PIL import Image
import json
import pandas as pd
```

```python
import os
from collections import Counter
counter = 0
patients = [str(i) for i in range(1, 49)]
resolutions = ["H", "L"]
splits = ["", " 2", " 3", " 4", " 5"]

for patient in patients:
    images = []

    for res in resolutions:
        for pth in splits:
            data_folder = f"/Volumes/One Touch/Data/Hematology/Large Leukemia Dataset/Leukemia_Attr/LeukemiaAttri_Dataset{pth}"

            for subset in ["train", "test"]:
                images_dir = os.path.join(data_folder, f"{res}_40X_C2/Images/{subset}")

                if not os.path.exists(images_dir):
                    continue

                for img in os.listdir(images_dir):
                    counter += 1
                    if img.startswith(f"{patient}_"):
                        images.append(img)

    counts = Counter(images)
    duplicates = [img for img, c in counts.items() if c > 1]

    print(patient, "total:", len(images), "unique:", len(set(images)), "duplicates:", len(duplicates))

    if duplicates:
        print("  duplicated files:", duplicates)
```
